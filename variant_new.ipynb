```r
variant <- fread("variant.csv")

# extract rare variants of the following classes: missense, stop, frameshift, splice
select_variant <- variant[variant$`SNP Type (l)` %in% c('Missense', 'Frameshift', 'Nonsense', 'Splicing'),]

# extract then the meta data
select_variant_meta <- select_variant[,1:16]
head(select_variant_meta)

# extract data for each cell line
variant_sample <- select_variant[,17:ncol(variant)]
variant_sample$Identifier <- select_variant$Identifier
nrow(variant_sample)
head(variant_sample)
```

```r
# use cutoff on the variant allele frequency to define presence of a variant in a given sample (start with > 0)
variant_sample_cutoff <- melt(variant_sample, variable.name = "sample(cell_line)", value.name = "frequency")
variant_sample_cutoff <- variant_sample_cutoff[variant_sample_cutoff$frequency > 0,]
head(variant_sample_cutoff)
nrow(variant_sample_cutoff)
```

```r
length(unique(variant_sample_cutoff$Identifier))
length(unique(variant_sample_cutoff$`sample(cell_line)`))
```

```r
# select variants with filtered frequency
select_variant_meta <- select_variant_meta[select_variant_meta$Identifier %in% unique(variant_sample_cutoff$Identifier),]
head(select_variant_meta)
nrow(select_variant_meta)
```

```r
# select the variants with dbSNPid
select_variant_meta_rs <- select_variant_meta[select_variant_meta$`dbSNP id (g)` != '-',]
head(select_variant_meta_rs)
nrow(select_variant_meta_rs)
```

```r
#if (!require("BiocManager", quietly = TRUE))
    #install.packages("BiocManager")
#options(timeout = 3000) 
#BiocManager::install("MafDb.gnomADex.r2.1.hs37d5")
# BiocManager::install("SNPlocs.Hsapiens.dbSNP155.GRCh37")
# BiocManager::install("SNPlocs.Hsapiens.dbSNP144.GRCh37")
library("MafDb.gnomADex.r2.1.hs37d5")

library(SNPlocs.Hsapiens.dbSNP155.GRCh37)
#SNPlocs.Hsapiens.dbSNP.20120608
```

```r
# if has dbSNPID, use ID to annotate MAF
snpdb <- SNPlocs.Hsapiens.dbSNP155.GRCh37
rng <- snpsById(snpdb, ids=unique(select_variant_meta_rs$`dbSNP id (g)`), ifnotfound="drop")
mafh5 <- MafDb.gnomADex.r2.1.hs37d5
annotate_rs <- gscores(mafh5, rng)                                            
```

```r
# MAF is sometimes NA, assume they are rare variants, just use a small value instead of NA
annotate_rs <- as.data.frame(annotate_rs)
nrow(annotate_rs)
nrow(annotate_rs[!is.na(annotate_rs$AF),])
nrow(annotate_rs[is.na(annotate_rs$AF),])
```

```r
# MAF is sometimes NA, assume they are rare variants, just use a small value instead of NA
annotate_rs <- as.data.frame(annotate_rs)

annotate_rs[is.na(annotate_rs$AF),]$AF <- 0.000001
nrow(annotate_rs)
head(annotate_rs)
nrow(annotate_rs[!is.na(annotate_rs$AF),])
```

```r
# for variants without ID, search with position instead
select_variant_meta_no_rs <- select_variant_meta[!select_variant_meta$`dbSNP id (g)` %in% annotate_rs$`RefSNP_id`,]
head(select_variant_meta_no_rs)
nrow(select_variant_meta_no_rs)
```

```r
length(unique(select_variant_meta_no_rs$Identifier))
```

```r
gr <- GRanges(seqnames = select_variant_meta_no_rs$`Chromosome (f)`, 
              ranges = IRanges(start = select_variant_meta_no_rs$`Start (f)`, end = select_variant_meta_no_rs$`End (f)`))
annotate_no_rs <- gscores(mafh5, gr)
annotate_no_rs <- as.data.frame(annotate_no_rs)
# again, # MAF is sometimes NA, assume they are rare variants, just use a small value instead of NA
annotate_no_rs[is.na(annotate_no_rs$AF),]$AF <- 0.000001

head(annotate_no_rs)
nrow(annotate_no_rs)
```

```r
# remove duplicates of the SNP
annotate_no_rs <- annotate_no_rs[!duplicated(annotate_no_rs),]
nrow(annotate_no_rs)
```

```r
# merge these annotated variants with the original meta data
select_variant_meta_rs_withAF <- merge(annotate_rs[,c(4,6)], select_variant_meta, by.y= 'dbSNP id (g)', by.x = 'RefSNP_id')
colnames(select_variant_meta_rs_withAF)[1] <- "dbSNP id (g)"
select_variant_meta_rs_withAF
```

```r
# merge these annotated variants with the original meta data
select_variant_meta_nors_withAF <- merge(annotate_no_rs[,c(1,2,3,6)], select_variant_meta,  by.y= c('Chromosome (f)', 'Start (f)', 'End (f)'), by.x = c('seqnames', 'start', 'end'))
colnames(select_variant_meta_nors_withAF)[1:3] <- c('Chromosome (f)', "Start (f)","End (f)")
select_variant_meta_nors_withAF
```
